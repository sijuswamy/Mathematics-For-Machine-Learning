## Vectors in `python`

The set $\mathbb{R}^n$ is the set of all n-tuples of real numbers. In set notation this is $\mathbb{R}^n=\{(x_1,x_2,x_3,⋯,x_n):x_1,x_2,x_3,\ldots ,x_n\in \mathbb{R}\}$. For example, the set $\mathbb{R}^3$ represents the set of real triples, $(x,y,z)$ coordinates, in three-dimensional space.

## Norm of a vector
The norm of a vector is a measure of its length. There are many ways of defining the length of a vector depending on the metric used (i.e., the distance formula chosen). The most common is called the $L_2$ norm, which is computed according to the distance formula you are probably familiar with from grade school. The $L_2$ norm of a vector $v$ is denoted by $∥v∥_2$ and 

$$\Vert v \Vert_{2} = \sqrt{\sum_i v_i^2}$$

This is sometimes also called Euclidian length and refers to the “physical” length of a vector in one-, two-, or three-dimensional space. The $L_1$ norm, or “Manhattan Distance,” is computed as

$$\Vert v \Vert_{1} = \sum_i |v_i|$$

and is named after the grid-like road structure in New York City. In general, the p-norm, $L_p$, of a vector is

$$\Vert v \Vert_{p} = \sqrt[p]{(\sum_i v_i^p)}$$

The $L_\infty$ norm is the p-norm, where $p=\infty$. The $L_\infty$ norm is written as $||v||_\infty$ and it is equal to the maximum absolute value in $v$.

## Python code for calculating norms

All the linear algebraic operations in python are defined in `numpy.linalg` library. So we have to import this library to implement vector/ matrix operations in python.

```python
import numpy.linalg as la
new_vector = vector_row.T
print("The Transposed vector\n",new_vector)
norm_1 = la.norm(new_vector, 1)
norm_2 = la.norm(new_vector, 2)
norm_inf = la.norm(new_vector, np.inf)
print('L_1 norm is: %.1f'%norm_1)
print('L_2 norm is: %.1f'%norm_2)
print('L_inf norm  is: %.1f'%norm_inf)
```

### `Dot` product
> The dot product of two vectors is the sum of the product of the respective elements in each vector and is denoted by $\cdot$, and $v\cdot w$ is read “v dot w.” This operation can be defined as $v\cdot w = \sum_{i = 1}^{n} v_iw_i$.

### Angle between two vectors
>Angle between two vectors can be calculated from the following relationship
$$v \cdot w = \Vert v \Vert_{2} \Vert w \Vert_{2} \cos{\theta}$$
So the angle between $v$ and $w$ is calculated as

$$\theta=\cos^{-1}\biggl(\dfrac{v\cdot w}{||v||_2\times ||w||_2}\biggr)$$
 An example is given bellow:





```python
v = np.array([[10, 9, 3]])
w = np.array([[2, 5, 12]])
theta = \
    np.arccos(np.dot(v, w.T)/(la.norm(v)*la.norm(w)))
print(theta)
```

## Creating augmented matrix using `numpy`

It is common to use the augmented matrix while solving system of linear equations. `concantenate` function from `numpy` library is used for this purpose. Suppose $A=\begin{pmatrix}1&1&0\\0&1&0\\1&0&1\end{pmatrix}$ and $y=\begin{pmatrix}1\\2\\1\end{pmatrix}$. Then  $[A:y]=\begin{pmatrix}1&1&0&1\\0&1&0&2\\1&0&1&1\end{pmatrix}$

```python
A = np.array([[1,1,0],
              [0,1,0],
              [1,0,1]])
print('Rank of A:', la.matrix_rank(A))
y = np.array([[1], [2], [1]])
A_y = np.concatenate((A, y), axis = 1)
print('Augmented matrix:\n', A_y)
print("Rank of Augmented matrix is:",la.matrix_rank(A_y))
```

### The motivation behind eigen values and eigen vectors in linear algebra

The motivation behind the eigenvalues and eigenvectors is that, it helps us to understand the characteristics of the linear transformation, thus make things easy. We know that a vector x can be transformed to a different vector by multiplying $A - Ax$. The effect of the transformation represents a scale of the length of the vector and/or the rotate of the vector. The above equation points out that for some vectors, the effect of transformation of $Ax$ is only scale (stretching, compressing, and flipping). The eigenvectors are the vectors have this property and the eigenvalues λ′s are the scale factors. Let’s look at the following example.

>Goal: Plot the vector `x = [[1], [1]]` and the vector `b=Ax`, where `A = [[2, 0], [0, 1]]`

```python
import matplotlib.pyplot as plt# for plotting purpose
plt.style.use('seaborn-poster')
%matplotlib inline

def plot_vect(x, b, xlim, ylim):## definition of plotting function
    '''
    function to plot two vectors, 
    x - the original vector
    b - the transformed vector
    xlim - the limit for x
    ylim - the limit for y
    '''
    plt.figure(figsize = (10, 6))
    plt.quiver(0,0,x[0],x[1],\
        color='k',angles='xy',\
        scale_units='xy',scale=1,\
        label='Original vector')
    plt.quiver(0,0,b[0],b[1],\
        color='g',angles='xy',\
        scale_units='xy',scale=1,\
        label ='Transformed vector')
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```

```python
A = np.array([[2, 0],[0, 1]])

x = np.array([[1],[1]])
b = np.dot(A, x)
plot_vect(x,b,(0,3),(0,2))
```

We can see from the generated figure that the original vector x is rotated and stretched longer after transformed by A. Let’s try to do the same exercise with a different vector `[[1], [0]]`

```python
x = np.array([[1], [0]])
b = np.dot(A, x)

plot_vect(x,b,(0,3),(-0.5,0.5))
```

Now we can see that with this new vector, the only thing changed after the transformation is the length of the vector, it is stretched. The new vector is `[[2], [0]]`, therefore, the transform is $$Ax=2x$$ Hence the vector `[[1],[0]]` is an eigen vector. Now try the vector `[[0],[1]]`.

```python
x = np.array([[0], [1]])
b = np.dot(A, x)

plot_vect(x,b,(-3,3),(-3,3))
```

## SVD Example of Dimensionality Reduction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

# read image in grayscale
img = cv2.imread('albert_einstein.jpg', 0)
print("image Resolution:",img.shape)
# obtain svd
U, S, V = np.linalg.svd(img)

# inspect shapes of the matrices
print(U.shape, S.shape, V.shape)

# plot images with different number of components
comps = [638, 500, 400, 300, 200, 5]

plt.figure(figsize = (16, 8))
for i in range(6):
  low_rank = U[:, :comps[i]] @ np.diag(S[:comps[i]]) @ V[:comps[i], :]
  if(i  == 0):
     plt.subplot(2, 3, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("Original Image with n_components =" + str(comps[i]))
  else:
     plt.subplot(2, 3, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("n_components =" + str(comps[i]))
```
